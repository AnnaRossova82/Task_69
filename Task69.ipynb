```python
(train_ds, val_ds, test_ds), metadata = tfds.load(
    'cifar10',
    split=['train[:80%]', 'train[80%:90%]', 'train[90%:]'],
    with_info=True,
    as_supervised=True,
)
```

```python
get_label_name = metadata.features['label'].int2str
```

```python
image, label = next(iter(test_ds))
plt.imshow(image)
plt.title(get_label_name)
```

```python
def visualize(original, augmented):
    fig = plt.figure()
    plt.subplot(1,2,1)
    plt.title('Original image')
    plt.imshow(original)
    plt.axis("off")
 
    plt.subplot(1,2,2)
    plt.title('Augmented image')
    plt.imshow(augmented)
    plt.axis("off")
```

```python
flipped = tf.image.flip_left_right(image)
visualize(image, flipped)
```

```python
saturation = tf.image.adjust_saturation(image, 4)
visualize(image, saturation)
```

```python
rotated = tf.image.rot90(image)
visualize(image, rotated)
```

```python
crop = tf.image.central_crop(image, central_fraction=0.4)
visualize(image, crop)
```

```python
brightness = tf.image.adjust_brightness(image, 0.2)
visualize(image, brightness)
```
